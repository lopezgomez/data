```
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/lopezgomez/data/master/NVDA.csv'

# Read the CSV file into a DataFrame
data = pd.read_csv(url, sep=',')

# Check if the required columns are in the DataFrame
if 'High' not in data.columns or 'Date' not in data.columns: raise ValueError("CSV file must contain 'X' and 'Y' columns")

# Plot the data
plt.figure(figsize=(8, 6))
plt.plot(data['Date'], data['High'], marker='o', linestyle='-', color='b', alpha=0.7)
plt.title('Plot of High by Date')
plt.xlabel('Date')
plt.ylabel('High')
plt.grid(True)
plt.show()
```
